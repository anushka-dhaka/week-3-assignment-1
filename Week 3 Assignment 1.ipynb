```
# Q1 Dataset overview
import pandas as pd

df = pd.read_csv("agriculture_yield_dataset.csv")

print("Shape:", df.shape)
print("Columns:")
print(df.columns.tolist())

print("First 10 Records:")
print(df.head(10))
```

```
#Q2. Data Types and Missing Values
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())
```

```
#Q3. Descriptive Statistics
# Mean values
mean_values = df.mean(numeric_only=True)
print("Mean values:")
print(mean_values)

# Standard deviation values
std_values = df.std(numeric_only=True)
print("\nStandard deviation values:")
print(std_values)

# Feature with highest mean
print("\nFeature with highest mean:")
print(mean_values.idxmax(), "=", mean_values.max())

# Feature with highest standard deviation
print("\nFeature with highest standard deviation:")
print(std_values.idxmax(), "=", std_values.max())
```

```
#Q4. Distribution Analysis
import matplotlib.pyplot as plt

# rainfall_mm
plt.figure(figsize=(6,4))
plt.hist(df['rainfall_mm'], bins=20)
plt.title('Distribution of Rainfall')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Frequency')
plt.show()

# temperature_c
plt.figure(figsize=(6,4))
plt.hist(df['temperature_c'], bins=20)
plt.title('Distribution of Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.show()

# fertilizer_kg
plt.figure(figsize=(6,4))
plt.hist(df['fertilizer_kg'], bins=20)
plt.title('Distribution of Fertilizer Usage')
plt.xlabel('Fertilizer (kg)')
plt.ylabel('Frequency')
plt.show()

# yield_ton_per_hectare
plt.figure(figsize=(6,4))
plt.hist(df['yield_ton_per_hectare'], bins=20)
plt.title('Distribution of Crop Yield')
plt.xlabel('Yield (ton/hectare)')
plt.ylabel('Frequency')
plt.show()
```

```
#Q7. Yield Distribution
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(df['yield_ton_per_hectare'], bins=20)

plt.title('Distribution of Crop Yield')
plt.xlabel('Yield (ton per hectare)')
plt.ylabel('Frequency')

plt.show()
```

```
#Q8. Scatter Plot Analysis
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.scatter(df['rainfall_mm'], df['yield_ton_per_hectare'])

plt.title('Rainfall vs Crop Yield')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Yield (ton/hectare)')

plt.show()
```

```
plt.figure(figsize=(6,4))
plt.scatter(df['fertilizer_kg'], df['yield_ton_per_hectare'])

plt.title('Fertilizer vs Crop Yield')
plt.xlabel('Fertilizer (kg)')
plt.ylabel('Yield (ton/hectare)')

plt.show()
```

```
#Q9. Correlation Analysis
corr_matrix = df.corr(numeric_only=True)

print(corr_matrix)
```

```
#heatmap
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.imshow(corr_matrix, aspect='auto')
plt.colorbar()

plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)

plt.title("Correlation Heatmap")
plt.show()
```

```
#Identify the top three features most correlated with crop yield.
yield_corr = corr_matrix['yield_ton_per_hectare'].sort_values(ascending=False)

print(yield_corr)
```

```
top3 = yield_corr.drop('yield_ton_per_hectare').head(3)

print("Top 3 features correlated with yield:")
print(top3)
```

```
#Q10. Group-Based Analysis
crop_yield = df.groupby('crop_type')['yield_ton_per_hectare'].mean()

print("Average Yield by Crop Type:")
print(crop_yield)
```

```
soil_yield = df.groupby('soil_type')['yield_ton_per_hectare'].mean()

print("\nAverage Yield by Soil Type:")
print(soil_yield)
```

```
print("Crop with Highest Average Yield:")
print(crop_yield.idxmax(), "-", crop_yield.max())

print("\nSoil Type with Highest Average Yield:")
print(soil_yield.idxmax(), "-", soil_yield.max())
```

```
crop_yield.plot(kind='bar', title='Average Yield by Crop Type')
plt.ylabel('Yield (ton/hectare)')
plt.show()

soil_yield.plot(kind='bar', title='Average Yield by Soil Type')
plt.ylabel('Yield (ton/hectare)')
plt.show()
```

```
#Q11. Feature Encoding
print(df.select_dtypes(include='object').columns)
```

```
df_encoded = pd.get_dummies(df, columns=['crop_type', 'soil_type'])

print(df_encoded.head())
```

```
print(df_encoded.head())
```

```
#questin 12
# Input features
X = df_encoded.drop('yield_ton_per_hectare', axis=1)

# Target variable
y = df_encoded['yield_ton_per_hectare']

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```
print("Input Features:")
print(X.columns)

print("\nTarget Variable:")
print(y.name)
```

```
#question 13
from sklearn.model_selection import train_test_split

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)

# Display shapes
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```
#question 14
from sklearn.linear_model import LinearRegression

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```
print("Intercept:", model.intercept_)

coefficients = pd.DataFrame({
    'Feature': X_train.columns,
    'Coefficient': model.coef_
})

print("\nCoefficients:")
print(coefficients)
```

```
highest_positive = coefficients.loc[coefficients['Coefficient'].idxmax()]

print("\nFeature with Highest Positive Coefficient:")
print(highest_positive)
```

```
coefficients = coefficients.sort_values(
    by='Coefficient',
    ascending=False
)

print(coefficients)
```
